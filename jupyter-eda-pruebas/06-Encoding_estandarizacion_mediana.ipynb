```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv("../datos/04-bikes_mediana.csv", index_col= 0).reset_index(drop= True)
df.head()
```

```python
# Según lo que vimos en visualización, eliminamos la columna "temp" por redundante
df.drop("temp", axis=1, inplace= True)
```

### Probamos primero estas opciones
- **Year** : Le damos más peso al año 2019
- **Workingday** : Le damós más peso a los días laborables (1)
- **Weathersit** : El encoding sería 1 - 4, para el 2 le ponemos un 2 y para el 3 le ponemos un 0
- **Holiday** : Le damos más peso a los días que no son festivos (0)
- **Season** : Hay que hacerlo. Verano 2, primavera y otoño 1, winter 0
- **mnth** enero: 0, febrero, marzo, abril, noviembre y diciembre: 1, junio, julio, agosto, septiembre, octubre: 2
- **Weekday** : Vamos probando difentes pesos

```python
map_yr = {0:1, 1:2}
df["yr"] = df["yr"].map(map_yr)
```

```python
map_workingday = {0:1, 1:2}
df["workingday"] = df["workingday"].map(map_workingday)
```

```python
map_weathersit = {3:0, 2:2, 1:4}
df["weathersit"] = df["weathersit"].map(map_weathersit)
```

```python
map_holiday = {0:7, 1:0}
df["holiday"] = df["holiday"].map(map_holiday)
```

```python
map_season = {"winter":0, "autumn":1, "spring":1, "summer":2}
df["season"] = df["season"].map(map_season)
```

```python
mapa_mnth = {1: 0, 2:0, 3: 1, 4: 1, 5: 2, 6:2, 7:2, 8:2, 9:2, 10:2, 11:1, 12:1}
df["mnth"] = df["mnth"].map(mapa_mnth)
```

```python
mapa_weekday = {1: 0, 2:1, 3: 1, 4: 2, 5: 2, 6:2, 7:1}
df["weekday"] = df["weekday"].map(mapa_weekday)
```

```python
df_est = df.copy()
df_est.head(1)
```

## ESTANDARIZACION

```python
# Creamos un dataframe con las columnas numéricas: atemp, hum y windspeed
num = df_est[["atemp", "hum", "windspeed"]]
num.head(2)
```

```python
# Estandarizamos las columnas con StandardScaler y sustituimos las columnas numéricas originales por las estandarizadas
st = StandardScaler()
x_tranformadas = pd.DataFrame(st.fit_transform(num))
x_tranformadas.columns = num.columns
df_est[x_tranformadas.columns] = x_tranformadas
df_est.head()
```

```python
df_est.to_csv("../datos/10-bikes_enc_est_mediana.csv")
```
